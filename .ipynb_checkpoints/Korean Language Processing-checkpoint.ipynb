```python
def build_vocab(data):
    flatten = lambda d: [token for sent in d for token in sent] # 이중 리스트 구조를 하나의 리스트 구조로 만들어준다.
    vocab = {}
    vocab['<unk>'] = 0
    vocab['<pad>'] = 1
    for token in flatten(list(zip(*data))[0]):
        if vocab.get(token) is None:
            vocab.setdefault(token, len(vocab))
    return vocab
    
vocab = build_vocab(sample_data)
#vocab = build_vocab(data)
```

```python
def numericalize(sent, vocab):
    temp= []
    for token in sent:
        if vocab.get(token) is None:
            temp.append(vocab['<unk>'])
        else:
            temp.append(vocab.get(token))
    return temp

numerical_data = [(numericalize(sent, vocab), label) \
                  for sent, label in sample_data]
#print(sample_data)
print(numerical_data[0])
```

```python
print(vocab)
```

```python
import torch
from collections import Counter
```

```python
#Bag of Words

num_docs = len(numerical_data)
print(num_docs)
num_vocab = len(vocab)
print(num_vocab)
term_matrix = torch.zeros(num_docs, num_vocab) # zeros()는 파라미터에 해당하는 사이즈의 0으로 채워진 텐서를 리턴

for i, (sent, label) in enumerate(numerical_data):# index와 함께 표시할때 enumerate이용
    tokens, cnts = zip(*list(Counter(sent).items()))
    term_matrix[i, torch.LongTensor(tokens)] = torch.FloatTensor(cnts)
    
print(term_matrix.size())
print(term_matrix.eq(0).sum())
# 차원이 너무 커져서 데이터의 특성을 표현해내기가 어렵다.
```

```python
print("Before Length of Sentences:")
print([len(sent) for sent, _ in numerical_data])
max_len = max([len(sent) for (sent, _) in numerical_data])

for sent, _ in numerical_data:
    if len(sent) < max_len:
        sent += [vocab['<pad>']] * (max_len - len(sent))
print("After Length of Sentences: ")
print([len(sent) for sent, _ in numerical_data])
print(numerical_data[0])
```

```python
from torchtext.data import Field

TEXT = Field(sequential=True,
             use_vocab=True,
             tokenize=tokenizer.morphs,
             lower=True,
             batch_first=True)
LABEL = Field(sequential=False,
             use_vocab=False,
             preprocessing=lambda x: int(x),
             batch_first=True,
              is_target=True)
ID = Field(sequential=False,
          use_vocab=False,
          is_target=False)
```

```python
from torchtext.data import TabularDataset

dataset = TabularDataset(path='./data/nsmc/ratings.txt',
                        format='tsv',
                        fields=[('id', ID), ('text', TEXT), ('label', LABEL)],
                        skip_header=True)
```

```python
TEXT.build_vocab(dataset)
print('Total vocabulary: {}'.format(len(TEXT.vocab)))
print('Token for "<unk>": {}'.format(TEXT.vocab.stoi['<unk>']))
print('Token for "<pad>": {}'.format(TEXT.vocab.stoi['<pad>']))
```

```python
from torchtext.data import Iterator

device = "cuda" if torch.cuda.is_available() else "cpu"
data_loader = Iterator(dataset=dataset,
                      batch_size=3,
                      device=device)

for batch in data_loader:
    break

print(batch.text.size(), batch.label.size())
print(batch.text)
```

```python
import torch
from torch.utils.data import Dataset, DataLoader

class CustomDataset(Dataset):
    def __init__(self, data_path): # 
        import numpy as np
        with open(data_path, "r", encoding="utf-8") as file:
            data = file.read().splitlines()
            data = np.array([line.split("\t") for line in data], dtype=np.float)
        self.x = data[:, :-1]
        self.t = data[:, -1]

    def __getitem__(self, index):
        return (self.x[index], self.t[index])
    def __len__(self):
        return len(self.x)
    def custom_collate_fn(self, data):
        x, t = list(zip(*data))
        return (torch.FloatTensor(x), torch.LongTensor(t))

example_dataset_path = "./data/example_dataset.tsv"
custom_dataset = CustomDataset(data_path=example_dataset_path)
train_loader = DataLoader(dataset=custom_dataset,
    batch_size=64,
    collate_fn=custom_dataset.custom_collate_fn,
    shuffle=True)

for x, y in train_loader:
    print("Size of mini-batch x: {}, t: {}".format(x.size(), y.size()))
    break
```
